# Surfaces

A **parameterized surface** is the image of a function

$\vec r:\RR^2 \to \RR^3$. We often write 

$$\vec r(u,v) = x(u,v) \,\vec i + y(u,v) \,\vec j + z(u,v) \,\vec k. $$

We look at three types:

  - graph of function of 2 variables. e.g., $z = f(x,y)$
  
  $$\vec r(u,v) = u \,\vec i + v \,\vec j + f(u,v) \,\vec k. $$

  - surface of revolution. e.g.,  $y = g(x)$ rotated about $x$-axis
  
  $$\vec r(u,v) = u \,\vec i + g(u)\cos v \,\vec j + g(u)\sin v \,\vec k. $$

  - part of sphere of radius $R$
  
$$\vec r(u,v) = R\sin u\cos v\,\vec i +  R\sin u \sin v\,\vec j +  R\cos u\,\vec k$$

## Example

Find the surface area of the part of the plane $z=x+2$ inside the cylinder $x^2+y^2=1$.

```python
fig= plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122,projection='3d')
u = v = np.linspace(-1,1,12)
r = lambda u,v: np.array([u,v,u+2])

ax1.set_xlabel('$u$',fontsize=18)
ax1.set_ylabel('$v$',fontsize=18)

ax2.set_xlabel('$x$',fontsize=18)
ax2.set_ylabel('$y$',fontsize=18)
ax2.set_zlabel('$z$',fontsize=18)

for i in u:
    ax1.plot(i*np.ones_like(v),v*sqrt(1-i**2),'b')
#     ax2.plot(i*np.ones_like(v),v*sqrt(1-i**2),'b',alpha=.5)
    ax2.plot(*r(i*np.ones_like(v),v*sqrt(1-i**2)),'b')
for i in v:
    ax1.plot(u*sqrt(1-i**2),i*np.ones_like(u),'r')
#     ax2.plot(u*sqrt(1-i**2),i*np.ones_like(u),'r',alpha=.5)
    ax2.plot(*r(u*sqrt(1-i**2),i*np.ones_like(u)),'r')

ax2.view_init(30,-135)
    
u,v = np.meshgrid(u,v)
# ax2.plot_wireframe(*r(u,v),color=('r','b'),rcount=10,ccount=10);
```

#### Solution

Parameterize as the graph of a function.

$$\vec r(u,v) = u \,\vec i +  v\,\vec j +  (u+2)\,\vec k$$

for $-1 \leq u \leq 1$ and $-\sqrt{1 - u^2} \leq v \leq \sqrt{1 - u^2}$ (that's a disk of radius 1).

$$ \vec r_u = \langle 1,0,1\rangle $$
$$ \vec r_v = \langle 0,1,0\rangle $$
$$ \vec r_u \times \vec r_v = \langle -1,0,1\rangle $$

$$\text{surface area} = \iint_\Omega dS = \int_{-1}^1 \int_{-\sqrt{1 - u^2}}^{\sqrt{1 - u^2}}\sqrt 2 \,dv\,du =\pi\sqrt 2$$

## Example

Find the surface area of the part of the plane $z=x+2$ inside the cylinder $x^2+y^2=1$.

```python
fig= plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122,projection='3d')
u = np.linspace(0,1,77)
v = np.linspace(0,2*pi,77)
r = lambda u,v: np.array([u*cos(v),u*sin(v),u*cos(v)+2])
ax1.set_xlabel('$u$',fontsize=18)
ax1.set_ylabel('$v$',fontsize=18)

ax1.set_xlim((0,3))

ax2.set_xlabel('$x$',fontsize=18)
ax2.set_ylabel('$y$',fontsize=18)
ax2.set_zlabel('$z$',fontsize=18)

for i in np.linspace(0,1,12):
    ax1.plot(i*np.ones_like(v),v,'b')
#     ax2.plot(i*cos(v),i*sin(v),0,'b',alpha=.5)
    ax2.plot(*r(i*np.ones_like(v),v),'b')
for i in np.linspace(0,2*pi,12):
    ax1.plot(u,i*np.ones_like(u),'r')
#     ax2.plot(u*cos(i),u*sin(i),0,'r',alpha=.5)
    ax2.plot(*r(u,i*np.ones_like(u)),'r')

ax2.view_init(30,-135)
    
u,v = np.meshgrid(u,v)
# ax2.plot_wireframe(*r(u,v),color=('r','b'),rcount=10,ccount=10);
```

#### Solution

Use cylindrical coordinates.

$$\vec r(u,v) = u\cos v \,\vec i +  u\sin v\,\vec j +  (u\cos v+2)\,\vec k$$

for $0 \leq u \leq 1$ and $0\leq v \leq 2\pi$ (that's a disk of radius 1).

$$ \vec r_u = \langle \cos v,\sin v, \cos v\rangle $$
$$ \vec r_v = \langle -u\sin v, u \cos v, -u\sin v \rangle $$
$$ \vec r_u \times \vec r_v = \langle -u,0,u\rangle $$

$$\text{surface area} = \iint_\Omega dS = \int_{0}^1 \int_{0}^{2\pi}u\sqrt 2 \,dv\,du =\pi\sqrt 2$$

### Example

Suppose a triangular chip is modeled as the piece of the plane $x+ y+ z = 1$ in the first octant with density $\rho(x,y,z) = 2-z$ (in mass per unit area). Find the mass of the chip. 

```python

@interact
def _(angle=(-96,186,6)):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(projection='3d')
    u = linspace(0,1)
    v = linspace(0,1)
    u,v = meshgrid(u,v)
    ax.plot_surface(u,(1-u)*v,1-u - (1-u)*v,color=(.1,1,.4,.5))
    ax.plot_wireframe(u,(1-u)*v,1-u - (1-u)*v,color='k',rcount=10,ccount=10)
    ax.view_init(39,angle)
    ax.set_zlim((0,1));
    flush_figures();
```

#### Solution

Parameterize as the graph of a function.

$$\vec r(u,v) = u \,\vec i +  v\,\vec j +  (1-u-v)\,\vec k$$

for $0 \leq u \leq 1$ and $0\leq v \leq 1-u$.

$$ \vec r_u = \langle 1,0,-1\rangle $$
$$ \vec r_v = \langle 0,1,-1\rangle $$
$$ \vec r_u \times \vec r_v = \langle 1,1,1\rangle $$

$$\text{mass} = \iint_\Omega \rho\,dS = $$ $$\int_{0}^1 \int_{0}^{1-u} \rho(x(u,v),y(u,v),z(u,v))|\vec r_u\times \vec r_v| \,dv\,du $$

$$=\int_{0}^1 \int_{0}^{1-u} 2 - (1 - u - v) \sqrt 3 \,dv\,du = \frac{5 \sqrt{3}}{6}$$

### Example

You are a scientist ar NORAD and get a tip that a UFO will be landing tonight "somewhere in the northern hemisphere". What is the expected latitude of its landing site?

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
u = linspace(0,pi/2)
v = linspace(0,2*pi)
u,v = meshgrid(u,v)
ax.plot_surface(sin(u)*cos(v),sin(u)*sin(v),cos(u),color=(.1,.4,1,.5))
ax.plot_wireframe(sin(u)*cos(v),sin(u)*sin(v),cos(u),color='k',rcount=10,ccount=10)
ax.set_zlim((0,2));
```

#### Solution

Call the northern hemisphere $N$. We compute the average latitude $$\bar\phi = \frac{\iint _N \phi \,dS}{\iint_N\,dS}$$

We will use the parametrization from spherical coordinates (taking the radius to be 1).

$$\vec r(u,v) = \sin u \cos v \,\vec i + \sin u \sin v \,\vec j + \cos u \,\vec k $$ 

for $0\leq u \leq \frac\pi2$ and $0 \leq v \leq 2\pi$.

This is a classical computation that follows, but must be done a few times in one's life.

$$\vec r_u =  \cos u \cos \,\vec i +  \cos u \sin v\,\vec j -\sin u  \,\vec k$$
$$\vec r_v = -\sin u \sin v \,\vec i +  \sin u \cos v\,\vec j $$

$$\vec r_u\times \vec r_v = \sin^2 u\cos v \,\vec i + \sin^2 u\sin v  \,\vec j + \sin u \cos u \,\vec k $$
$$= \sin u \vec r$$ which makes sense. The position is orthogonal to the surface. The magnitude is thus $\sin u$, smaller near the north pole and largest near the equator. Further we know the surface area is $2\pi$ (half that of unit sphere). So, 

$$\bar\phi = \frac{1}{2\pi}\int_0^{\pi/2} \int_0^{2\pi} u\sin u \,dv\,du = 1$$

The integrand is such because $u = \phi$ in this parameterization. 

This is neat. It means the average latitude in the north hemisphere is 1 radian from the pole.
